b) Get app url by regular expression using functions from __[re module](https://docs.python.org/3/library/re.html?highlight=re#module-re)__.

A useful online regular expression check.
__[Check your regular expression first](https://regex101.com)__.

We go though all the HTML we retrieved earlier and extract app urls with the given regex expression, getting all unique app urls and collecting them into a larger set.

```python
import re
appreg = r'href=\"(/store/apps/details.*?)\"'
appre = re.compile(appreg)

all_app_urls = []
for s in sources:
    app_url_list = re.findall(appre,s)
    app_url_list = list(set(app_url_list))
    all_app_urls += app_url_list
    #print(len(app_url_list))
    #print(app_url_list)
    
all_app_urls = list(set(all_app_urls))
print(len(all_app_urls))
print(all_app_urls)
```

This approach just yields ~750 unique app urls. We do another round using those app urls to find similar apps in order to go above 1000 apps.

```python
second_round_urls = []

for url in all_app_urls[1:200]:
    try:
        x = urllib.request.urlopen('https://play.google.com/' + url).read().decode('utf-8')
        app_url_list = re.findall(appre,x)
        app_url_list = list(set(app_url_list))
        second_round_urls += app_url_list
    except:
        print('error on', url)
        
second_round_urls = list(set(second_round_urls))
print(len(second_round_urls))
print(second_round_urls)
```

Great! That yielded another ~1100 apps. Next, we concatenate that set with the app urls we found earlier.

```python
final_app_urls = list(set(all_app_urls + second_round_urls))
print(len(final_app_urls))
print(final_app_urls)
```

We are now at ~1700 urls and can continue with the next step.

c) Access specific webpage to get description of each app and then store the description in files.

For each url we want to exctract the description so we store them in a dictionary, mapping descriptions to app names, and save them all in a json file. 

```python
import json
descdict = dict()
for u in range(0,len(final_app_urls)):
    #Get web content
    appname = final_app_urls[u].split('=')[1]
    url = 'https://play.google.com'+final_app_urls[u]
    content = urllib.request.urlopen(url).read().decode('utf-8')
    #Regex
    descreg = "itemprop=\"description.*?\">.*?<div jsname=\".*?\">.*?</div>"
    descre = re.compile(descreg)
    description = re.findall(descre,content)[0]
    #Store in dictionary
    descdict[appname]=description

#Save to file
with open('descriptions.json', 'w') as fp:
    json.dump(descdict, fp)
```

d) Preprocess text using NLP techniques from __[nltk module](http://www.nltk.org/py-modindex.html)__ or spaCy.

Using nltk.download(ID) to get the corpora if it is not downloaded before. __[nltk corpora](http://www.nltk.org/nltk_data/)__

For each app we process the description in this order:
1. Remove html tags
2. Tokenization
3. Removal of stop words
4. Stemming

```python
import nltk
from nltk.stem import *
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
from bs4 import BeautifulSoup

#Create some nltk objects
nltk.download('stopwords')
stemmer = SnowballStemmer("english")
nltk.download('punkt')
documents = descdict.items()

#Lists of names and descriptions
corpus = []
names = []
for k,v in documents:
    #Remove html tags
    soup = BeautifulSoup(v, 'html.parser')
    text = soup.get_text()
    #Tokenize
    tokens = nltk.word_tokenize(text)
    #Remove stop words
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    #Stemming
    for i in range(0,len(tokens)):
        tokens[i]=re.sub(r'([^\s\w]|_)+', '', tokens[i])
        tokens[i]=stemmer.stem(tokens[i])
        
    text = ' '.join(tokens)
    corpus.append(text)
    names.append(k)
    
```

...)Compute tdidf 
eg. Using functions from __[scikit-learn module](http://scikit-learn.org/stable/modules/classes.html)__. TfidfVectorizer is used for converting a collection of raw documents to a matrix of TF-IDF features.
#### You can also build the tfidf matrix with other library or your own algorithm

```python
from sklearn.feature_extraction.text import TfidfVectorizer
transvector = TfidfVectorizer()

tfidf1 = transvector.fit_transform(corpus)
#print(tfidf1.toarray())
#print(transvector.get_feature_names())

```

```python
from scipy import spatial
def search(term, tfidf, corpus):
    tfidf_search = transvector.transform([term])
    tfidf_search_vector = tfidf_search.toarray()[0]
    
    matches = []
    
    for vector, sentence, name in zip(tfidf.toarray(), corpus, names):
        similarity = 1 - spatial.distance.cosine(tfidf_search_vector, vector)
        matches.append((similarity,name, sentence))
        
    matches.sort(reverse=True)
    return matches[:3]

res = search('Facebook like friend', tfidf1, corpus)

print(res)
```
